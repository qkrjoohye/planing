---## Part 2: 📖 WBS, RACI, 리스크 관리### 2.1 WBS (Work Breakdown Structure)복잡한 프로젝트를 관리 가능한 작업 단위로 분해.- **계층적 분해**: 프로젝트 → 단계 → 결과물 → 작업 패키지- **100% 규칙**: 하위 항목의 합 = 상위 항목 100%- **결과물 중심**: "회의 참석"이 아닌 "요구사항 명세서 작성"### 2.2 RACI 매트릭스| 역할 | 의미 | 규칙 ||------|------|------|| **R** (Responsible) | 실제 작업 수행자 | 복수 가능 || **A** (Accountable) | 최종 책임자 | **반드시 1명!** || **C** (Consulted) | 자문 제공자 | 남발 금지 || **I** (Informed) | 통보 대상자 | 꼭 필요한 사람만 |### 2.3 리스크 관리 프로세스1. **식별**: 브레인스토밍, 체크리스트2. **평가**: 확률 × 영향 = 리스크 점수3. **대응**: 회피 / 완화 / 전가 / 수용4. **모니터링**: 트리거 점검, 레지스터 업데이트### 2.4 리스크 대응 전략 4가지| 전략 | 정의 | 적용 상황 | 예시 ||------|------|----------|------|| **회피** | 리스크 원인 제거 | 높은 확률 × 높은 영향 | 고위험 기술 대체 || **완화** | 확률/영향 감소 | 중간 수준 리스크 | 테스트 강화 || **전가** | 제3자에게 이전 | 보험/외주 가능 | 전문 업체 외주 || **수용** | 리스크 감수 | 낮은 리스크/불가피 | 예비비 확보 |---

---## Part 3: 📖 몬테카를로 일정 시뮬레이션### 3.1 크리티컬 패스 분석프로젝트에서 **가장 긴 경로** = 크리티컬 패스. 이 경로의 작업이 지연되면 전체 프로젝트 기한 지연.### 3.2 확률적 일정 추정각 작업의 기간을 **삼각분포**로 모델링:- **최소(a)**: 모든 것이 순조로울 때- **최빈(m)**: 가장 가능성 높은 기간- **최대(b)**: 문제 발생 시 최악의 경우$$\text{기대값} = \frac{a + 4m + b}{6}, \quad \text{표준편차} = \frac{b - a}{6}$$### 3.3 몬테카를로 시뮬레이션10,000회 시뮬레이션으로 **완료 일정의 확률 분포** 도출.- P50 = 50% 확률 완료 기간- P90 = 90% 확률 완료 기간 (보수적 추정)- 버퍼 = P90 - 평균### 3.4 AI 기반 실행 지원- 일정 충돌 자동 탐지- 크리티컬 패스 실시간 재계산- 자연어 질의: "가장 위험한 작업은?"---

```python
# ========================================# 몬테카를로 일정 시뮬레이션# ========================================# 프로젝트 작업 정의 (삼각분포: min, mode, max)tasks_mc = {    'T1_Requirements':   {'min': 5,  'mode': 7,  'max': 12, 'deps': []},    'T2_SystemDesign':   {'min': 6,  'mode': 8,  'max': 14, 'deps': ['T1_Requirements']},    'T3_UIDesign':       {'min': 4,  'mode': 6,  'max': 10, 'deps': ['T1_Requirements']},    'T4_BackendDev':     {'min': 20, 'mode': 30, 'max': 45, 'deps': ['T2_SystemDesign']},    'T5_FrontendDev':    {'min': 15, 'mode': 25, 'max': 35, 'deps': ['T2_SystemDesign', 'T3_UIDesign']},    'T6_AIIntegration':  {'min': 12, 'mode': 20, 'max': 30, 'deps': ['T4_BackendDev']},    'T7_UnitTest':       {'min': 5,  'mode': 7,  'max': 10, 'deps': ['T4_BackendDev', 'T5_FrontendDev', 'T6_AIIntegration']},    'T8_IntegrationTest':{'min': 7,  'mode': 10, 'max': 15, 'deps': ['T7_UnitTest']},    'T9_Deployment':     {'min': 3,  'mode': 5,  'max': 8,  'deps': ['T8_IntegrationTest']},}def simulate_schedule(tasks, n_sims=10000):    """몬테카를로 일정 시뮬레이션"""    np.random.seed(42)    task_names = list(tasks.keys())    total_durations = []    critical_counts = {t: 0 for t in task_names}    for _ in range(n_sims):        durations = {}        finish_times = {}        for name, info in tasks.items():            dur = np.random.triangular(info['min'], info['mode'], info['max'])            durations[name] = dur            start = max([finish_times[d] for d in info['deps']], default=0)            finish_times[name] = start + dur        total = max(finish_times.values())        total_durations.append(total)        # 크리티컬 패스 식별 (역추적)        last_task = max(finish_times, key=finish_times.get)        critical_counts[last_task] += 1    return np.array(total_durations), critical_countsdurations, crit_counts = simulate_schedule(tasks_mc)print("📊 몬테카를로 일정 시뮬레이션 결과 (10,000회)")print("=" * 55)print(f"  평균 기간:  {np.mean(durations):.1f}일")print(f"  표준편차:   {np.std(durations):.1f}일")print(f"  P50 (중앙값): {np.percentile(durations, 50):.1f}일")print(f"  P75:         {np.percentile(durations, 75):.1f}일")print(f"  P90:         {np.percentile(durations, 90):.1f}일")print(f"  P95:         {np.percentile(durations, 95):.1f}일")print(f"\n  90% 신뢰수준 버퍼: {np.percentile(durations, 90) - np.mean(durations):.1f}일")
```

```python
# ========================================# 시각화: 완료 기간 분포 + 목표 달성 확률# ========================================fig = make_subplots(rows=1, cols=2,    subplot_titles=['Project Duration Distribution', 'Target Date Achievement Probability'])# 히스토그램fig.add_trace(go.Histogram(x=durations, nbinsx=50, marker_color='#3498db',                           opacity=0.7, name='Duration'), row=1, col=1)percentiles = {'P50': 50, 'P75': 75, 'P90': 90, 'P95': 95}colors_p = {'P50': '#2ecc71', 'P75': '#f1c40f', 'P90': '#e67e22', 'P95': '#e74c3c'}for label, pct in percentiles.items():    val = np.percentile(durations, pct)    fig.add_vline(x=val, line_dash='dash', line_color=colors_p[label],                  annotation_text=f'{label}={val:.0f}d', row=1, col=1)# 목표별 달성 확률targets = [80, 85, 90, 95, 100, 105, 110]probs = [(durations <= t).mean() * 100 for t in targets]bar_colors = ['#e74c3c' if p < 30 else '#f1c40f' if p < 70 else '#2ecc71' for p in probs]fig.add_trace(go.Bar(x=[f'{t}d' for t in targets], y=probs,                     marker_color=bar_colors,                     text=[f'{p:.0f}%' for p in probs], textposition='auto',                     name='Probability'), row=1, col=2)fig.add_hline(y=80, line_dash='dash', line_color='green', row=1, col=2,              annotation_text='80% target')fig.update_layout(height=400, width=1000, title='Monte Carlo Schedule Simulation',                  showlegend=False)fig.update_xaxes(title_text='Duration (days)', row=1, col=1)fig.update_xaxes(title_text='Target Duration', row=1, col=2)fig.update_yaxes(title_text='Frequency', row=1, col=1)fig.update_yaxes(title_text='Achievement Probability (%)', row=1, col=2)fig.show()print("💡 목표 일정별 달성 확률:")for t, p in zip(targets, probs):    rec = '비현실적' if p < 10 else '도전적' if p < 30 else '달성가능' if p < 70 else '안정적' if p < 90 else '보수적'    print(f"  {t}일: {p:.1f}% ({rec})")
```
