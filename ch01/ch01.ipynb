## 1.1 기획의 정의

### 기획이란?

- **기획(Planning)**은 다음의 과정이다:
1. **바람직한 미래 상태**(목표)를 설정하고

## 1.2 기획 vs 분석 vs 계획 vs 전략

### 핵심 구분

| 개념 | 정의 | 초점 | 핵심 질문 | 산출물 |
| --- | --- | --- | --- | --- |
| **분석(Analysis)** | 현상을 이해하는 것 | 과거/현재 | 무슨 일이 왜 일어났나? | 보고서, 인사이트 |
| **기획(Planning)** | 행동 계획을 설계하는 것 | 미래 | 무엇을 어떻게 할 것인가? | 계획서 |
| **계획(Plan)** | 기획의 문서화된 결과물 | 미래 | - | 문서 |
| **전략(Strategy)** | 장기적 방향 설정 | 미래 | 어디로 갈 것인가? | 비전, 방향 |

### 시각적 비교

````python
# 시각화: 개념 비교

concepts = ['Analysis', 'Planning', 'Plan', 'Strategy']
time_focus = [1, 3, 3, 4]  # 1=Past, 2=Present, 3=Near Future, 4=Far Future
action_level = [1, 4, 3, 2]  # 1=Understanding, 2=Direction, 3=Document, 4=Design
scope = [2, 3, 2, 4]  # 1=Narrow, 4=Broad

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=[time_focus[0], action_level[0], scope[0]],
    theta=['Time Focus', 'Action Level', 'Scope'],
    fill='toself',
    name='Analysis'
))

fig.add_trace(go.Scatterpolar(
    r=[time_focus[1], action_level[1], scope[1]],
    theta=['Time Focus', 'Action Level', 'Scope'],
    fill='toself',
    name='Planning'
))

fig.add_trace(go.Scatterpolar(
    r=[time_focus[3], action_level[3], scope[3]],
    theta=['Time Focus', 'Action Level', 'Scope'],
    fill='toself',
    name='Strategy'
))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 5])),
    title='Comparison: Analysis vs Planning vs Strategy',
    height=450
)

fig.show()

print("핵심 인사이트:")
print("  - 분석(Analysis): 과거에 초점, 이해 중심")
print("  - 기획(Planning): 미래에 초점, 행동 설계 중심")
print("  - 전략(Strategy): 먼 미래에 초점, 방향 설정 중심")
````

### Mintzberg의 통찰

> **"기획은 분석(ANALYSIS)에 관한 것이고, 전략은 종합(SYNTHESIS)에 관한 것이다."**
> 
> - 분석: 전체를 부분으로 나누는 것
> - 종합: 부분들을 새로운 전체로 결합하는 것

이것이 "전략적 기획(Strategic Planning)"이 **모순어법(oxymoron)**이 될 수 있는 이유다!

---

### Mintzberg의 핵심 주장과 함의

#### 1. 핵심 주장 (*The Rise and Fall of Strategic Planning*, 1994)

> **"전략은 계획되는 것이 아니라, 출현(emerge)하는 것이다."**

| 개념 | Mintzberg의 정의 | 사고 방식 |
|---|---|---|
| **기획(Planning)** | 전략을 실행 단계로 운영화하는 것 | 분석적 — 쪼개기 |
| **전략(Strategy)** | 경험·학습·통찰에서 발현되는 방향 | 종합적 — 합치기 |

- **형식화의 함정:** 전략을 프로세스로 만들면 창의성이 죽고, "계획 = 전략"이라는 착각이 생긴다
- **창발적 전략(Emergent Strategy):** 진짜 전략은 현장의 행동 패턴에서 자연스럽게 발현된다 (예: 소니 워크맨, 혼다의 미국 진입)

---

#### 2. 이 과정에 주는 함의 — 패러다임의 전환

Mintzberg 비판의 핵심은 전통적 기획이 **미래를 예측하고 처방하려 한다**는 것이다.

> 전통적 기획: "2030년에 X를 달성하기 위해 Y를 한다"

그러나 전략은 예측에서 나오지 않는다. 그렇다면 기획은 무엇을 해야 하는가?

**답: 미래를 계획하지 말고, 의사결정 시스템을 설계하라.**

이것이 이 과정이 제시하는 **SAPD(State-based Adaptive Policy Design)** 의 핵심 발상이다:

| 전통적 기획 | SAPD |
|---|---|
| **결과**를 계획한다 | **의사결정 함수**를 설계한다 |
| 미래를 **예측**한다 | 상태 변화에 **반응**한다 |
| 전략을 **문서화**한다 | 전략이 **출현하는 구조**를 만든다 |
| 계획 → 실행 (선형) | State(t) → Decision → Response → State(t+1) (순환) |

Mintzberg는 "전략을 계획할 수 없다"고 했다. SAPD는 이에 동의하며 이렇게 답한다:

> **"그러므로 전략을 계획하지 않는다. 대신, 조건이 변할 때 좋은 판단이 생성되는 의사결정 아키텍처를 설계한다."**

$$State(t) \rightarrow Decision(t) = f(State(t)) \rightarrow Response \rightarrow State(t+1) \rightarrow \cdots$$

이 과정에서 배우는 모든 도구는 이 순환 루프의 **구성 요소**다:

- **인과추론(5-6장):** "무엇이 무엇을 야기하는가?" → Decision 함수 $f$의 근거
- **베이지안 사고(9장):** "우리는 얼마나 확실한가?" → State 관측의 불확실성 정량화
- **시스템 다이내믹스(6장):** "피드백 루프는 어떻게 작동하는가?" → Response → State(t+1) 메커니즘
- **시나리오 기획(10장):** "상태가 어떻게 바뀔 수 있는가?" → 복수의 State(t+1) 경로 탐색
- **몬테카를로 시뮬레이션(11장):** "결과의 분포는?" → Decision의 리스크 평가

**핵심:** 이 도구들은 고정된 계획을 만들기 위한 것이 아니라, **적응적 의사결정 시스템을 구축하기 위한 인프라**다. 이것이 전통적 기획 교육과 이 과정의 근본적 차이다.

## 1.3 기획의 유형

### 세 가지 주요 유형

## 1.4 6단계 기획 프로세스

### 개요

```
1. 문제 정의     →  "문제가 무엇인가?"
        ↓
2. 원인 분석     →  "왜 그런 일이 일어나는가?"
        ↓
3. 대안 개발     →  "무엇을 할 수 있는가?"
        ↓
4. 의사결정      →  "무엇을 해야 하는가?"
        ↓
5. 실행         →  "어떻게 실행할 것인가?"
        ↓
6. 학습         →  "무엇을 배웠는가?"
        ↓
   (1단계로 복귀)
```

### 핵심 인사이트: 기획은 선형이 아니라 반복적이다!

### 아인슈타인의 지혜

> **"문제를 해결하는 데 1시간이 주어진다면, 55분은 문제를 이해하는 데, 5분은 해결책을 찾는 데 쓰겠다."**

**시사점:** 문제 정의(1단계)가 가장 중요한 단계다!

## 1.5 전통적 프레임워크가 충분하지 않은 이유

### 일반적인 프레임워크

| 프레임워크 | 목적 | 한계 |
| --- | --- | --- |
| SWOT | 내부/외부 분석 | 정적, 주관적 |
| PEST | 거시 환경 스캔 | 서술적, 인과관계 부재 |
| 5 Forces | 산업 경쟁 분석 | 안정적 구조 가정 |
| BCG Matrix | 포트폴리오 분석 | 지나치게 단순화 |

### 문제점

전통적 프레임워크는:

1. **인과관계를 확립하지 못한다** - 무엇이 무엇을 야기하는가?
2. **불확실성을 무시한다** - 가정이 틀리면?
3. **정적이다** - 변화하는 조건에 적응하지 못한다

### 해결책: 현대적 접근법

| 접근법 | 무엇을 추가하는가 | 해당 장 |
| --- | --- | --- |
| **인과추론** | "X가 정말로 Y를 야기하는가?" | 5-6장 |
| **베이지안 사고** | "우리는 얼마나 확실한가?" | 9장 |
| **시스템 다이내믹스** | "것들이 어떻게 상호작용하는가?" | 6장 |
| **시나리오 기획** | "만약에...?" | 10장 |
| **몬테카를로 시뮬레이션** | "결과의 분포는 어떤가?" | 11장 |

````python
# 시각화: 전통적 vs 현대적 접근법

approaches = pd.DataFrame({
    'Dimension': ['Causality', 'Uncertainty', 'Dynamics', 'Quantification', 'Adaptability'],
    'Traditional (SWOT, PEST)': [1, 1, 1, 2, 1],
    'Modern (Causal, Bayesian)': [5, 5, 4, 5, 4]
})

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=approaches['Traditional (SWOT, PEST)'],
    theta=approaches['Dimension'],
    fill='toself',
    name='Traditional',
    line=dict(color='gray')
))

fig.add_trace(go.Scatterpolar(
    r=approaches['Modern (Causal, Bayesian)'],
    theta=approaches['Dimension'],
    fill='toself',
    name='Modern',
    line=dict(color='blue')
))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 6])),
    title='Traditional vs Modern Planning Approaches',
    height=450
)

fig.show()

print("현대적 접근법이 중요한 이유:")
print("  1. 인과성: SWOT은 강점이 성공을 야기하는지 말해주지 않는다")
print("  2. 불확실성: 점추정은 위험을 숨긴다 (9장 참고)")
print("  3. 역동성: 정적 분석은 피드백 루프를 놓친다")
````
