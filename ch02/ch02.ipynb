## Part 1: 📖 SWOT 분석 이론

### 1.1 SWOT란 무엇인가?

**SWOT 분석**은 조직이나 프로젝트의 전략적 위치를 파악하기 위한 프레임워크입니다.

1960년대 스탠퍼드 연구소(SRI)의 **앨버트 험프리(Albert Humphrey)**가 기업 전략 연구에서 개발한 이래, 경영학에서 가장 널리 쓰이는 분석 도구 중 하나가 되었습니다.

SWOT는 네 가지 요소의 머리글자입니다:

| | **긍정적 요인** | **부정적 요인** |
|---|---|---|
| **내부 요인** (우리가 통제 가능) | **S**trengths (강점) | **W**eaknesses (약점) |
| **외부 요인** (우리가 통제 불가) | **O**pportunities (기회) | **T**hreats (위협) |

#### 쉽게 이해하기: 취업 준비 비유

여러분이 데이터 분석가로 취업을 준비한다고 합시다:

- **강점(S)**: Python 능숙, 통계학 수업 A+, 공모전 수상 경력
- **약점(W)**: 영어 회화 부족, 인턴 경험 없음, 포트폴리오 미흡
- **기회(O)**: AI/빅데이터 채용 급증, 정부 디지털 전환 정책, 부트캠프 지원금
- **위협(T)**: 전공자 외 경쟁자 유입(비전공 전환), AI가 일부 업무 자동화, 경기 침체

핵심은 **내부 요인**과 **외부 요인**을 구분하는 것입니다:
- **내부**: 내가 노력하면 바꿀 수 있는 것 (강점을 강화하거나, 약점을 보완하거나)
- **외부**: 내가 통제할 수 없지만, 대비하거나 활용할 수 있는 것

---

### 1.2 SWOT 매트릭스 구조

좋은 SWOT 분석을 위한 **질문 가이드**:

| 요소 | 핵심 질문 | 예시 (카페 창업) |
|------|-----------|------------------|
| **강점(S)** | 우리가 잘하는 것은? 경쟁자보다 나은 점은? | 바리스타 자격증 보유, 대학가 입지, 낮은 임대료 |
| **약점(W)** | 부족한 자원은? 개선할 점은? | 초기 자본 부족, 마케팅 경험 없음, 좁은 매장 |
| **기회(O)** | 활용할 수 있는 외부 변화는? | 재택근무 → 동네 카페 수요 증가, SNS 마케팅 비용 저렴 |
| **위협(T)** | 주의해야 할 외부 위험은? | 프랜차이즈 카페 진출, 원두 가격 상승, 최저임금 인상 |

#### SWOT 작성 시 흔한 실수

1. **내부/외부 혼동**: "시장이 성장한다"는 강점이 아니라 **기회**입니다
2. **너무 추상적**: "품질이 좋다" → "원두 직접 로스팅으로 신선도 유지" (구체적으로!)
3. **나열만 하고 끝**: SWOT는 분석의 **시작**이지 끝이 아닙니다 → TOWS 전략으로 연결해야 합니다

---

````python
# ============================================================
# SWOT 매트릭스 시각화: 대학 주변 카페 창업 사례
# ============================================================

# SWOT 요인 정의
swot_data = {
    'Strengths (강점)': [
        '바리스타 자격증 보유',
        '대학가 유동인구 많음',
        '상대적으로 낮은 임대료',
        '젊은 감각의 인테리어 가능'
    ],
    'Weaknesses (약점)': [
        '초기 자본금 부족',
        '마케팅·경영 경험 없음',
        '매장 면적 15평으로 협소',
        '브랜드 인지도 제로'
    ],
    'Opportunities (기회)': [
        '재택근무 확대 → 동네 카페 선호',
        'SNS 마케팅 비용 저렴',
        '스페셜티 커피 수요 증가',
        '소상공인 창업 지원금 확대'
    ],
    'Threats (위협)': [
        '반경 500m 내 프랜차이즈 3곳',
        '원두 가격 30% 상승 추세',
        '최저임금 인상 → 인건비 부담',
        '경기 침체 → 소비 위축 우려'
    ]
}

# 2x2 매트릭스 시각화
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=['💪 Strengths (강점)', '⚠️ Weaknesses (약점)',
                    '🌟 Opportunities (기회)', '🔥 Threats (위협)'],
    vertical_spacing=0.12, horizontal_spacing=0.12
)

colors = {'S': '#2196F3', 'W': '#FF9800', 'O': '#4CAF50', 'T': '#F44336'}
positions = {'Strengths (강점)': (1,1), 'Weaknesses (약점)': (1,2),
             'Opportunities (기회)': (2,1), 'Threats (위협)': (2,2)}
color_keys = {'Strengths (강점)': 'S', 'Weaknesses (약점)': 'W',
              'Opportunities (기회)': 'O', 'Threats (위협)': 'T'}

for category, items in swot_data.items():
    r, c = positions[category]
    y_vals = list(range(len(items)-1, -1, -1))
    fig.add_trace(
        go.Bar(
            x=[1]*len(items), y=[f"{i+1}. {item}" for i, item in enumerate(items)],
            orientation='h',
            marker_color=colors[color_keys[category]],
            showlegend=False,
            text=[f"{i+1}. {item}" for i, item in enumerate(items)],
            textposition='inside',
            insidetextanchor='start'
        ),
        row=r, col=c
    )

fig.update_xaxes(visible=False)
fig.update_yaxes(visible=False)
fig.update_layout(
    title='SWOT 분석: 대학 주변 카페 창업',
    height=500, width=900,
    template='plotly_white'
)
fig.show()

print("\n📌 관찰 포인트:")
print("   - 내부 요인(위)은 우리가 바꿀 수 있고, 외부 요인(아래)은 대비만 가능합니다.")
print("   - 이 SWOT만으로는 '그래서 뭘 해야 하나?'에 답하기 어렵습니다.")
print("   - → 다음에 배울 TOWS 매트릭스가 이 문제를 해결합니다.")
````

### 1.3 SWOT 활용 전략: TOWS 매트릭스

SWOT를 작성한 것만으로는 부족합니다. **"그래서 어떤 전략을 세울 것인가?"**에 답하려면 **TOWS 매트릭스**가 필요합니다.

TOWS는 SWOT의 네 요소를 **교차 조합**하여 전략을 도출합니다:

| | **기회 (O)** | **위협 (T)** |
|---|---|---|
| **강점 (S)** | **SO 전략** (공격적 전략) | **ST 전략** (다각화 전략) |
| | 강점으로 기회를 극대화 | 강점으로 위협을 최소화 |
| **약점 (W)** | **WO 전략** (방향 전환 전략) | **WT 전략** (방어적 전략) |
| | 약점을 보완하여 기회 활용 | 약점과 위협을 동시에 최소화 |

#### 카페 창업 TOWS 예시

| 전략 | 조합 | 구체적 전략 |
|------|------|-------------|
| **SO** | 바리스타 자격 × 스페셜티 수요 증가 | 직접 로스팅한 스페셜티 커피를 시그니처 메뉴로 출시 |
| **WO** | 마케팅 경험 없음 × SNS 마케팅 저렴 | 인스타그램 릴스로 로스팅 과정 공유, 팔로워 확보 |
| **ST** | 대학가 입지 × 프랜차이즈 진출 | 학생 맞춤 가격 정책(텀블러 할인, 단골 적립) 도입 |
| **WT** | 초기 자본 부족 × 원두 가격 상승 | 소상공인 지원금 활용 + 소량 다품종 로스팅으로 재고 최소화 |

**핵심**: SWOT는 "현재 상태 파악", TOWS는 "행동 전략 도출"입니다.

---

### 1.4 SWOT의 장점과 한계

#### 장점
- **직관적**: 누구나 쉽게 이해하고 참여 가능 (회의에서 바로 활용)
- **범용적**: 기업, 비영리, 개인, 정책 등 어디에나 적용 가능
- **소통 도구**: 팀원 간 전략적 상황에 대한 공통 인식 형성
- **빠른 착수**: 별도의 데이터 없이도 전문가 의견으로 시작 가능

#### 한계
- **주관적**: 같은 조직이라도 작성자에 따라 결과가 달라질 수 있음
- **우선순위 부재**: 10개 강점이 나열되어도 "가장 중요한 것"을 알기 어려움
- **정량화 어려움**: "강점이 얼마나 강한지"를 숫자로 표현하지 않음
- **정적(Static)**: 특정 시점의 스냅샷 → 환경 변화를 반영하지 못함
- **외부 분석의 체계성 부족**: O와 T를 어떻게 찾을지 가이드가 없음 → **PESTLE이 이 문제를 해결**

## Part 2: 📖 PEST/PESTLE 분석 이론

### 2.1 PEST 분석이란?

SWOT의 한계 중 하나는 **외부 환경(O, T)을 체계적으로 찾는 방법이 없다**는 것입니다.  
"기회가 뭐가 있지?"라고 막연히 생각하면, 중요한 것을 놓치기 쉽습니다.

**PEST 분석**은 외부 환경을 **4가지 카테고리**로 나누어 빠짐없이 스캔하는 프레임워크입니다.

| 요인 | 영문 | 쉬운 비유 | 분석 질문 예시 |
|------|------|-----------|----------------|
| **P** (정치적) | Political | 게임의 규칙 | 정부 정책, 규제, 세금, 보조금은? |
| **E** (경제적) | Economic | 지갑 사정 | 경기, 금리, 환율, 물가는 어떤가? |
| **S** (사회적) | Social | 사람들의 취향 | 인구 구조, 라이프스타일, 가치관 변화는? |
| **T** (기술적) | Technological | 새로운 도구 | 신기술, 자동화, 디지털화 추세는? |

#### 예시: 대학 주변 카페의 PEST

- **P**: 소상공인 창업지원금 확대, 1회용품 규제 강화
- **E**: 원두 수입가 상승(환율), 소비 심리 위축, 최저임금 인상
- **S**: MZ세대 '카공족' 문화, 비건/디카페인 수요 증가, 1인 가구 증가
- **T**: 키오스크·무인 결제, 배달앱 연동, SNS 마케팅 플랫폼

이렇게 4가지 렌즈로 보면, 막연한 "기회/위협"이 체계적으로 정리됩니다.

---

### 2.2 PEST → PESTLE 확장

2000년대 이후 **ESG(환경·사회·지배구조)**, **규제 강화**, **기후 위기** 등이 중요해지면서 PEST에 두 가지가 추가되었습니다:

| 추가 요인 | 영문 | 설명 | 왜 중요해졌나? |
|-----------|------|------|----------------|
| **L** (법적) | Legal | 법률, 노동법, 소비자보호법, 지식재산권 | GDPR(개인정보), 플랫폼 규제 등 법적 리스크 증가 |
| **E** (환경적) | Environmental | 기후변화, 탄소 규제, 폐기물, 에너지 | ESG 경영, 탄소중립 정책, 소비자의 친환경 요구 |

#### PEST vs PESTLE: 언제 뭘 쓸까?

| 상황 | 추천 도구 | 이유 |
|------|-----------|------|
| 빠른 환경 스캔이 필요할 때 | PEST | 4가지로 충분히 개괄 가능 |
| 규제·법률이 중요한 산업 (의료, 금융, 식품) | PESTLE | L 요인이 사업에 결정적 영향 |
| ESG·환경 이슈가 큰 산업 (에너지, 제조, 물류) | PESTLE | E 요인이 비용·평판에 직결 |
| 정책 분석, 공공 기획 | PESTLE | 법적·환경적 요인이 핵심 |

> 💡 이 수업에서는 더 포괄적인 **PESTLE**을 기본으로 사용합니다.

---

````python
# ============================================================
# PESTLE 6요인 레이더 차트: 한국 전기차 시장 사례
# ============================================================

categories = ['Political\n(정치적)', 'Economic\n(경제적)', 'Social\n(사회적)',
              'Technological\n(기술적)', 'Legal\n(법적)', 'Environmental\n(환경적)']

# 각 요인의 영향도 (1=낮음, 5=높음)
# 한국 전기차 시장 기준으로 평가
scores = [5, 4, 4, 5, 3, 5]

# 각 요인별 근거
reasons = [
    'P: 전기차 보조금, 충전 인프라 투자 정책',
    'E: 유가 변동, 배터리 원자재 가격, 소비자 구매력',
    'S: 친환경 소비 트렌드, 젊은 세대 전기차 선호',
    'T: 배터리 기술 혁신, 자율주행, 충전 속도 개선',
    'L: 배출가스 규제, 전기차 의무판매제, 안전 기준',
    'E(환경): 탄소중립 2050, 미세먼지 저감 정책'
]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=scores + [scores[0]],  # 닫힌 다각형
    theta=categories + [categories[0]],
    fill='toself',
    fillcolor='rgba(33, 150, 243, 0.2)',
    line=dict(color='#2196F3', width=2),
    name='영향도',
    text=reasons + [reasons[0]],
    hovertemplate='%{text}<br>영향도: %{r}/5<extra></extra>'
))

fig.update_layout(
    polar=dict(
        radialaxis=dict(visible=True, range=[0, 5], tickvals=[1,2,3,4,5]),
    ),
    title='PESTLE 분석: 한국 전기차 시장 (영향도 1~5)',
    height=500, width=600,
    template='plotly_white'
)
fig.show()

print("\n📌 해석 포인트:")
print("   - P(정치), T(기술), E(환경)이 모두 5점 → 전기차 시장은 정책·기술·환경이 동시에 추동")
print("   - L(법적)이 상대적으로 낮음 → 아직 규제 프레임이 완성 중")
print("   - 레이더 차트로 보면 '어떤 요인이 가장 큰 영향을 미치는지' 한눈에 파악 가능")
````

### 2.3 SWOT과 PESTLE의 관계

PESTLE과 SWOT는 **경쟁 관계가 아니라 보완 관계**입니다.

```
분석 흐름:

  ┌─────────────┐      ┌─────────────┐      ┌─────────────┐      ┌─────────────┐
  │  PESTLE      │ ──→  │  SWOT의     │ ──→  │  내부 분석   │ ──→  │  TOWS       │
  │  외부환경    │      │  O와 T 도출  │      │  S와 W 추가  │      │  전략 도출   │
  │  6요인 스캔  │      │             │      │             │      │             │
  └─────────────┘      └─────────────┘      └─────────────┘      └─────────────┘
```

#### 왜 이 순서가 좋은가?

1. **PESTLE 먼저**: 6가지 렌즈로 외부 환경을 빠짐없이 스캔
2. **O/T 도출**: PESTLE 결과에서 우리에게 유리한 것(O)과 불리한 것(T)을 구분
3. **S/W 추가**: 내부 역량을 솔직하게 평가
4. **TOWS 전략**: 네 요소를 교차 조합하여 구체적 행동 전략 수립

#### 전기차 시장 사례로 연결하면

| PESTLE 결과 | → SWOT 연결 |
|-------------|-------------|
| P: 정부 보조금 확대 | → **기회(O)**: 보조금 활용 시 가격 경쟁력 확보 |
| E: 배터리 원자재 가격 상승 | → **위협(T)**: 생산 원가 상승 압박 |
| T: 충전 기술 발전 | → **기회(O)**: 충전 불편 해소로 수요 확대 |
| L: 배출가스 규제 강화 | → **기회(O)**: 내연기관 퇴출 → 전기차 전환 가속 |

---

### 2.4 실무에서의 활용

SWOT와 PESTLE은 다양한 실무 상황에서 활용됩니다:

| 활용 분야 | PESTLE 역할 | SWOT 역할 | 산출물 |
|-----------|------------|-----------|--------|
| **사업계획서** | 시장 환경 분석 | 우리 사업의 경쟁력 평가 | 투자 유치 IR 자료 |
| **정책 보고서** | 정책 환경 스캐닝 | 정책 대안의 장단점 비교 | 정책 제안서 |
| **컨설팅 프로젝트** | 산업 분석 (외부) | 클라이언트 진단 (내부+외부) | 전략 보고서 |
| **취업 준비** | 채용 시장 트렌드 분석 | 나의 경쟁력 포지셔닝 | 자기소개서, 면접 대비 |
| **대학원 연구** | 연구 배경 (거시 환경) | 연구 갭(Gap) 식별 | 논문 서론 |

실무에서 발표할 때는 보통 이런 순서로 보고합니다:

1. **PESTLE로 큰 그림** 먼저 보여주고 ("시장 환경은 이렇습니다")
2. **SWOT로 우리 위치** 정리하고 ("이 환경에서 우리는 이런 상태입니다")
3. **TOWS로 전략** 제안 ("따라서 이런 전략을 추천합니다")

## Part 3: 📖 SWOT 정량화와 시각화

### 3.1 왜 정량화가 필요한가?

Part 1에서 SWOT의 한계 중 하나가 **"우선순위를 알 수 없다"**는 것이었습니다.

예를 들어, 카페 창업 SWOT에서 강점이 4개 나왔을 때:
- 바리스타 자격증 보유
- 대학가 유동인구 많음
- 상대적으로 낮은 임대료
- 젊은 감각의 인테리어 가능

**이 중 가장 중요한 강점은 무엇일까요?** 단순 나열로는 알 수 없습니다.

#### 정량화 방법: 중요도 × 점수

각 요인에 두 가지 점수를 부여합니다:

1. **중요도 (Weight)**: 이 요인이 성공에 얼마나 중요한가? (1~5점)
2. **평가 점수 (Rating)**: 현재 우리 상태가 얼마나 좋은가? (1~5점)
3. **가중 점수 = 중요도 × 평가 점수** → 이 값으로 우선순위 결정

| 강점 요인 | 중요도 | 평가 점수 | 가중 점수 |
|-----------|--------|-----------|----------|
| 바리스타 자격증 | 4 | 5 | **20** |
| 대학가 유동인구 | 5 | 4 | **20** |
| 낮은 임대료 | 3 | 4 | **12** |
| 젊은 인테리어 감각 | 2 | 3 | **6** |

→ 바리스타 자격증과 대학가 입지가 **동등하게 중요**하고, 인테리어 감각은 상대적으로 덜 중요

---

### 3.2 실습 시연: 대학 주변 카페 창업 SWOT 정량 분석

아래 코드는 카페 창업 SWOT의 모든 요인을 정량화하고 시각화합니다.  
코드를 **실행만** 하고, 결과를 함께 해석해 봅시다.

---

````python
# ============================================================
# SWOT 정량화 분석: 대학 주변 카페 창업
# ============================================================

# --- 데이터 준비 ---
swot_quant = pd.DataFrame({
    'Category': ['S','S','S','S', 'W','W','W','W', 'O','O','O','O', 'T','T','T','T'],
    'Factor': [
        # Strengths
        '바리스타 자격증', '대학가 유동인구', '낮은 임대료', '인테리어 감각',
        # Weaknesses
        '초기 자본 부족', '마케팅 경험 없음', '매장 면적 협소', '브랜드 인지도 없음',
        # Opportunities
        '동네 카페 선호 트렌드', 'SNS 마케팅 저비용', '스페셜티 수요 증가', '창업 지원금',
        # Threats
        '프랜차이즈 경쟁', '원두 가격 상승', '최저임금 인상', '소비 위축'
    ],
    'Weight': [4,5,3,2, 5,4,3,4, 4,3,4,3, 5,4,3,3],    # 중요도 (1~5)
    'Rating': [5,4,4,3, 5,3,2,4, 4,4,3,3, 4,3,3,2],    # 평가 점수 (1~5)
})

# 가중 점수 계산
swot_quant['Weighted'] = swot_quant['Weight'] * swot_quant['Rating']

# --- 카테고리별 색상 ---
cat_colors = {'S': '#2196F3', 'W': '#FF9800', 'O': '#4CAF50', 'T': '#F44336'}
cat_names = {'S': '강점(S)', 'W': '약점(W)', 'O': '기회(O)', 'T': '위협(T)'}

# --- 시각화 1: 카테고리별 가중 점수 막대 차트 ---
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=[f"{cat_names[c]} 가중 점수" for c in ['S','W','O','T']],
    vertical_spacing=0.15, horizontal_spacing=0.15
)

for idx, cat in enumerate(['S','W','O','T']):
    df_cat = swot_quant[swot_quant['Category'] == cat].sort_values('Weighted', ascending=True)
    r, c = divmod(idx, 2)
    fig.add_trace(
        go.Bar(
            x=df_cat['Weighted'], y=df_cat['Factor'],
            orientation='h',
            marker_color=cat_colors[cat],
            text=df_cat['Weighted'],
            textposition='outside',
            showlegend=False
        ),
        row=r+1, col=c+1
    )
    fig.update_xaxes(range=[0, 30], row=r+1, col=c+1)

fig.update_layout(
    title='SWOT 정량화: 카페 창업 요인별 가중 점수 (중요도 × 평가)',
    height=550, width=900,
    template='plotly_white'
)
fig.show()

# --- 시각화 2: 카테고리별 총점 비교 ---
summary = swot_quant.groupby('Category')['Weighted'].sum().reindex(['S','W','O','T'])

fig2 = go.Figure(go.Bar(
    x=[cat_names[c] for c in summary.index],
    y=summary.values,
    marker_color=[cat_colors[c] for c in summary.index],
    text=summary.values,
    textposition='outside'
))

fig2.update_layout(
    title='SWOT 카테고리별 총 가중 점수',
    xaxis_title='SWOT 카테고리',
    yaxis_title='총 가중 점수',
    height=400, width=600,
    template='plotly_white'
)
fig2.show()

# --- 해석 출력 ---
print("\n" + "="*60)
print("📊 정량화 SWOT 분석 결과 요약")
print("="*60)

for cat in ['S','W','O','T']:
    df_cat = swot_quant[swot_quant['Category'] == cat].sort_values('Weighted', ascending=False)
    top = df_cat.iloc[0]
    total = df_cat['Weighted'].sum()
    print(f"\n{cat_names[cat]} (총점: {total})")
    print(f"  → 최우선 요인: {top['Factor']} (가중 점수: {top['Weighted']})")

internal = summary['S'] - summary['W']
external = summary['O'] - summary['T']
print(f"\n📌 내부 역량 점수 (S-W): {internal:+d}  {'→ 내부 강점 우위' if internal > 0 else '→ 내부 약점 우위'}")
print(f"📌 외부 환경 점수 (O-T): {external:+d}  {'→ 기회 우위' if external > 0 else '→ 위협 우위'}")
print(f"\n💡 종합: {'공격적 SO 전략 추천' if internal > 0 and external > 0 else '상황에 맞는 복합 전략 필요'}")
````

## Part 5: 🔬 실습 - AI 활용 SWOT·PESTLE 분석

### 5.1 AI로 SWOT/PESTLE을 더 잘하는 법

지금까지 우리는 **직접 머리를 써서** SWOT과 PESTLE을 작성했습니다.  
이제 **AI(ChatGPT, Claude 등)**를 활용하면 이 과정이 어떻게 달라지는지 체험해 봅시다.

#### AI가 도와줄 수 있는 영역

| 영역 | 전통적 방식 | AI 활용 방식 | 개선 효과 |
|------|------------|-------------|----------|
| **정보 수집** | 직접 뉴스·보고서 검색 | AI가 트렌드·이슈 요약 | 시간 절약, 범위 확대 |
| **요인 도출** | 브레인스토밍 (경험 의존) | AI가 다양한 관점에서 요인 제시 | 누락 방지, 다양성 증가 |
| **교차 검증** | 동료 리뷰 (시간 소요) | AI에게 "빠진 것이 있나?" 질문 | 빠른 피드백 |
| **반론 생성** | Devil's Advocate 역할 필요 | AI가 반대 의견 제시 | 편향 감소 |
| **전략 연결** | TOWS 매트릭스 수동 작성 | AI가 SWOT→전략 초안 생성 | 초안 속도 향상 |

#### AI 활용 시 주의할 점 (Human-in-the-Loop)

1. **AI 결과는 초안이지 최종본이 아닙니다** → 반드시 검토·수정
2. **환각(Hallucination) 주의**: AI가 그럴듯하지만 틀린 정보를 만들 수 있음
3. **맥락 부족**: AI는 우리 조직 내부 사정을 모름 → 내부 요인(S, W)은 인간이 보완
4. **편향 가능**: AI 학습 데이터의 편향이 분석에 반영될 수 있음

> 💡 **핵심 원칙**: AI는 분석의 **속도와 범위**를 높여주고, 인간은 **판단과 검증**을 담당한다.

## 🎓 강의 마무리 및 핵심 요약

### 오늘 배운 내용

#### 1. SWOT 분석
- 내부(S, W) × 외부(O, T)의 2×2 매트릭스로 전략적 위치를 파악
- TOWS 매트릭스로 SO/WO/ST/WT **행동 전략**을 도출
- 주관적이고 정적이라는 한계가 있음

#### 2. PESTLE 분석
- 외부 환경을 P·E·S·T·L·E 6가지 렌즈로 체계적으로 스캔
- SWOT의 O(기회)와 T(위협)를 빠짐없이 도출하는 도구
- PESTLE → SWOT → TOWS 순서가 가장 효과적

#### 3. SWOT 정량화
- 중요도 × 평가 점수 = 가중 점수로 우선순위 결정
- S-W(내부 역량), O-T(외부 환경) 점수로 전략 방향 판단
- Python(pandas + plotly)으로 자동 계산 및 시각화

#### 4. AI 활용
- AI는 속도·범위·다양성을 높여주지만, 내부 맥락과 검증은 인간의 몫
- 최적의 방식은 **AI 초안 + 인간 검증의 하이브리드**

---

### 핵심 메시지

> **SWOT는 '현재 상태'를, PESTLE은 '외부 환경'을, TOWS는 '행동 전략'을 알려준다.**  
> **세 가지를 조합하면 체계적이고 실행 가능한 전략이 완성된다.**
