### 1.2 AI의 4가지 기여 영역

McKinsey(2024)에 따르면, AI는 기획의 4가지 핵심 활동을 증강한다:

| 기여 영역 | AI의 역할 | 활용 예시 | 한계 |
|-----------|-----------|-----------|------|
| **정보 수집** | 대규모 데이터 처리, 약신호 탐지 | 뉴스/보고서 자동 분석 | 맥락 이해 부족 |
| **인과 발견** | 데이터 기반 인과구조 추론 | DAG 자동 제안 | 전문가 검증 필수 |
| **시나리오 생성** | 미래 시나리오 서사 작성 | 4-시나리오 매트릭스 | 창의성 한계 |
| **일관성 검증** | 논리적 모순 탐지, 편향 경고 | 전략 스트레스 테스트 | 가치 판단 불가 |

### 1.3 시각화: AI 증강의 효과

````python
# AI 증강 전후 비교: 기획 프로세스 소요 시간
stages = ['Information<br>Gathering', 'Causal<br>Analysis', 'Scenario<br>Development', 
          'Decision<br>Making', 'Report<br>Writing']
traditional = [40, 25, 30, 20, 25]  # 전통적 방식 (시간)
ai_augmented = [10, 15, 12, 15, 5]  # AI 증강 방식 (시간)

fig = go.Figure()

fig.add_trace(go.Bar(
    name='Traditional',
    x=stages, y=traditional,
    marker_color='lightcoral',
    text=[f'{h}h' for h in traditional],
    textposition='outside'
))

fig.add_trace(go.Bar(
    name='AI-Augmented',
    x=stages, y=ai_augmented,
    marker_color='steelblue',
    text=[f'{h}h' for h in ai_augmented],
    textposition='outside'
))

fig.update_layout(
    title='Planning Process: Traditional vs AI-Augmented (Hours)',
    yaxis_title='Hours',
    barmode='group',
    height=450
)

fig.show()

total_trad = sum(traditional)
total_ai = sum(ai_augmented)
reduction = (1 - total_ai / total_trad) * 100
print(f"\n📊 총 소요 시간 비교:")
print(f"  전통적 방식: {total_trad}시간 ({total_trad/8:.0f}일)")
print(f"  AI 증강 방식: {total_ai}시간 ({total_ai/8:.0f}일)")
print(f"  시간 절감: {reduction:.0f}%")
print(f"\n💡 핵심: AI는 '정보 수집'과 '보고서 작성'에서 가장 큰 효과")
print(f"   반면 '의사결정'은 여전히 인간의 판단이 핵심")
````

### 1.4 정보 수집과 패턴 인식

기획의 첫 단계는 관련 정보를 수집하고 그 속에서 의미 있는 패턴을 찾는 것이다.

**전통적 방식의 한계:**
- 정보의 양이 기하급수적으로 증가
- 인간의 인지 능력으로 모든 정보를 처리하기 어려움
- **확증 편향(Confirmation Bias)**: 기존 관점과 일치하는 정보에만 주목

**AI의 기여:**
- 뉴스, 보고서, 학술 논문, 소셜 미디어에서 정보 자동 수집
- **약신호(Weak Signal)** 포착: 아직 주류가 아니지만 미래에 중요해질 조짐
- BCG(2025): AI 확장 적용 기업의 74%가 정보 수집에 AI 활용

**실무 활용 영역:**
1. 뉴스 및 보고서 자동 분석
2. 특허 및 학술 논문 분석
3. 소셜 미디어 감성 분석
4. 규제 환경 모니터링

### 1.5 인과관계 발견 지원

기획에서 가장 중요한 질문: **"무엇이 무엇에 영향을 미치는가?"**

| 유형 | 알고리즘 | 원리 | 장점 |
|------|----------|------|------|

| 딥러닝 기반 | CASTLE | 신경망 기반 학습 | 비선형 관계 포착 |

> **주의**: AI가 생성한 인과 그래프는 **가설**로서 취급해야 하며, 도메인 전문가의 검토가 필수적이다.

### 1.6 시뮬레이션과 시나리오 생성

### 1.7 의사결정 일관성 검증

AI는 의사결정의 **도전자(Challenger)** 역할을 수행한다:

### 2.7 간단 실습: 점 추정 vs 확률 분포

"시장 규모는 100억 원"이라는 점 추정이 왜 위험한지 시뮬레이션으로 확인한다.

- 평균: 100억 원
- 표준편차: 30억 원 (불확실성)
- 손익분기점: 80억 원

````python
# 점 추정 vs 확률 분포 비교
n_sim = 10000
market_sizes = np.random.normal(100, 30, n_sim)  # 평균 100억, 표준편차 30억
breakeven = 80  # 손익분기점

prob_success = (market_sizes > breakeven).sum() / n_sim
ci_5 = np.percentile(market_sizes, 5)
ci_95 = np.percentile(market_sizes, 95)

fig = go.Figure()

fig.add_trace(go.Histogram(
    x=market_sizes, nbinsx=50,
    name='Market Size Distribution',
    marker_color='lightblue', opacity=0.7
))

fig.add_vline(x=breakeven, line_dash='dash', line_color='red', line_width=3,
              annotation_text=f'Break-even (80B)<br>P(exceed): {prob_success:.1%}',
              annotation_position='top right')

fig.add_vline(x=100, line_dash='solid', line_color='green', line_width=3,
              annotation_text='Point Estimate (100B)',
              annotation_position='top left')

fig.add_vrect(x0=min(market_sizes), x1=breakeven, 
              fillcolor='red', opacity=0.1)

fig.update_layout(
    title=f'Market Size: Point Estimate vs Probability Distribution ({n_sim:,} simulations)',
    xaxis_title='Market Size (Billion KRW)',
    yaxis_title='Frequency',
    height=450, showlegend=False
)

fig.show()

print(f"📊 시뮬레이션 결과 ({n_sim:,}회):")
print(f"  - 평균: {market_sizes.mean():.1f}억 원")
print(f"  - 90% 신뢰구간: [{ci_5:.1f}억, {ci_95:.1f}억]")
print(f"  - 손익분기점(80억) 초과 확률: {prob_success:.1%}")
print(f"  - 손실 위험: {1 - prob_success:.1%}")
print(f"\n💡 '시장 규모는 100억'보다 '75% 확률로 80억 초과'가 정직한 보고")
print(f"   → 9장(베이지안), 11장(몬테카를로)에서 자세히 다룬다")
````

### 3.1 Human-in-the-Loop (HITL) 원칙

**정의**: AI 시스템의 운영, 감독, 의사결정에 인간이 적극적으로 참여하는 것

**왜 필수인가?**
- EU AI Act 제14조: 고위험 AI 시스템에 인간 감독 의무
- GDPR 제22조: 법적 효과 발생 결정의 완전 자동화 금지

#### HITL 5가지 핵심 원칙

1. **AI 출력은 제안이지 결정이 아니다**
2. **최종 판단과 책임은 인간에게 있다**
3. **AI의 한계를 인식하고 검증 프로세스를 갖춘다**
4. **중요한 결정일수록 인간 검토 단계를 강화한다**
5. **AI와 인간의 역할 경계를 명확히 정의한다**

#### HITL 구현 프레임워크

| 결정 유형 | AI 역할 | 인간 역할 | 검토 수준 |
|-----------|---------|-----------|----------|
| 정보 수집 | 자동 수행 | 품질 샘플링 | 낮음 |
| 분석/요약 | 초안 생성 | 검토 및 수정 | 중간 |
| 옵션 평가 | 대안 제시, 점수화 | 최종 선택 | 높음 |
| 전략 결정 | 시나리오 제시 | 판단 및 결정 | 매우 높음 |

#### 주의: HITL의 트레이드오프

- **Automation Bias**: AI 출력에 과도하게 의존
- **Algorithm Aversion**: 정확한 AI 출력마저 무시
- → "인간을 포함시키는 것"이 아니라 **"적절한 방식으로 포함시키는 것"**이 핵심

### 3.2 환각(Hallucination)과 편향

#### 환각이란?

LLM이 **"그럴듯하지만 사실이 아닌 내용을 생성하는 것"**

- **내재적 환각**: 입력과 모순되는 출력 생성
- **외재적 환각**: 입력에서 검증할 수 없는 정보 생성

#### 환각 완화 5가지 전략

1. **출처 요청**: "이 통계의 출처는 무엇인가?"
2. **교차 검증**: 다른 AI 모델이나 외부 DB와 비교
3. **RAG**: 신뢰할 수 있는 문서를 참조하여 답변 생성
4. **Chain-of-Verification**: AI에게 자신의 답변을 검증하도록 요청
5. **불확실성 표현 요청**: "확실하지 않은 부분은 그렇다고 명시하라"

#### 편향의 유형

| 편향 유형 | 설명 | 위험 |
|-----------|------|------|
| 지리적 편향 | 영어권(특히 미국) 중심 | 타 지역 분석 부정확 |
| 시간적 편향 | 학습 데이터 시점까지만 반영 | 최신 동향 누락 |
| 산업적 편향 | IT/금융 과다, 타 산업 빈약 | 특정 산업 분석 편향 |
| 관점적 편향 | 학습 데이터의 세계관 반영 | 특정 관점 과대 대표 |

|----------------|----------------|

| 일관된 기준의 평가 | 진정한 창의성과 혁신적 도약 |

### 쉽게 이해하기 3-1

#### AI의 한계와 HITL 원칙

AI는 빠르고 유용하지만 항상 맞는 답을 주지는 않는다.

- **환각(Hallucination)**은 그럴듯하지만 사실이 아닌 내용을 만들어내는 문제다.

- **RAG**는 외부 문서 검색 결과를 함께 사용해 답변의 근거를 보강하는 방식이다.

- **Human-in-the-Loop**는 중요한 판단에서 사람이 중간에 검토하고 승인해야 한다는 원칙이다.

핵심은 AI를 자동결정 기계가 아니라, 검토 가능한 보조도구로 써야 한다는 점이다.

---

### 핵심 정리 3-1

1. 환각은 AI가 사실이 아닌 내용을 그럴듯하게 만드는 현상이다.

2. RAG는 문서를 함께 찾아보게 해서 근거 없는 답변을 줄인다.

3. 중요한 의사결정은 반드시 사람이 검토해야 하며, 이것이 HITL의 핵심이다.

## 🎓 강의 마무리 및 핵심 요약

### 오늘 배운 내용

#### 1. AI의 4가지 기여 영역
- 정보 수집과 패턴 인식 (약신호 탐지)
- 인과관계 발견 지원 (DAG 생성)
- 시뮬레이션과 시나리오 생성 (확률적 미래 탐색)
- 의사결정 일관성 검증 (편향 탐지, 스트레스 테스트)

#### 2. 기획 단계별 AI 활용
- 문제 정의 → 인과 분석 → 미래 탐색 → 의사결정
- 단계별로 적합한 AI 도구가 다르다

#### 3. AI 활용의 원칙과 한계
- **HITL 원칙**: AI 출력은 제안, 최종 판단은 인간
- **환각과 편향**: 반드시 검증 프로세스 필요
- AI가 잘하는 것(데이터 처리)과 못하는 것(가치 판단) 구분

#### 4. AI 도구 3가지 층위
- LLM (분석 지원, 시나리오 생성)
- Python 라이브러리 (정량 분석, 시뮬레이션)
- AI 에이전트 (워크플로우 자동화)

---

### 핵심 메시지

> **AI는 기획자를 대체하는 것이 아니라 증강(Augment)한다.**
>
> AI는 데이터 처리, 패턴 인식, 시뮬레이션에서 인간의 한계를 보완하지만,  
> 조직 문화의 이해, 가치 판단, 최종 책임은 여전히 인간의 몫이다.  
> 좋은 기획자는 AI를 **도구로 활용하되 맹신하지 않는** 균형 감각을 갖춘 사람이다.
