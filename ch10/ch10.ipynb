# Part 1: 시나리오 플래닝의 본질

## 1.1 예측(Forecasting) vs 시나리오 플래닝

전통적 기획은 **단일 미래**를 예측하고 그에 맞춰 전략을 수립한다. 그러나 VUCA(Volatility, Uncertainty, Complexity, Ambiguity) 환경에서는 "가장 가능성 높은 미래"에 베팅하는 것이 위험하다.

| 구분 | 예측(Forecasting) | 시나리오 플래닝 |
|------|-------------------|----------------|
| 미래관 | 단일 미래 (확률 기반) | 복수의 가능한 미래 |
| 목적 | 정확한 예측 | 전략적 준비와 학습 |
| 불확실성 | 줄이려 함 | 활용하려 함 |
| 산출물 | 숫자 (매출, 성장률 등) | 서사(narrative) + 시사점 |
| 대표 사례 | 경제성장률 전망 | Shell 에너지 시나리오 |

## 1.2 시나리오 플래닝의 기원

Royal Dutch Shell은 1970년대 초 시나리오 플래닝을 통해 석유 위기를 사전에 대비한 것으로 유명하다. Pierre Wack이 이끈 Shell의 기획팀은 중동 정세 변화 가능성을 시나리오로 구성했고, 1973년 석유 파동 때 Shell은 경쟁사 대비 빠르게 대응할 수 있었다.

> **핵심 통찰**: 시나리오 플래닝의 가치는 "미래를 맞히는 것"이 아니라, "다양한 미래에 대비하는 조직 역량"을 구축하는 데 있다.

## 1.3 STEEP 프레임워크

시나리오 플래닝의 첫 단계는 **동인(Driving Forces)** 식별이다. STEEP 프레임워크는 외부 환경을 체계적으로 스캔하는 도구이다:

- **S**ocial (사회적): 인구 구조, 라이프스타일, 가치관 변화
- **T**echnological (기술적): 신기술 등장, 기술 성숙도, 디지털 전환
- **E**conomic (경제적): 경기 순환, 산업 구조, 무역 환경
- **E**nvironmental (환경적): 기후 변화, 자원 제약, 환경 규제
- **P**olitical (정치적): 정권 변화, 규제 방향, 국제 관계

각 동인에 대해 **불확실성(Uncertainty)**과 **영향도(Impact)**를 평가하여 핵심 불확실성을 식별한다.

# Part 2: 불확실성-영향도 매트릭스

## 2.1 동인 분류 프레임워크

STEEP 분석으로 식별된 동인들을 **불확실성(Uncertainty)**과 **영향도(Impact)** 두 축으로 평가한다. 이를 통해 4개 유형으로 분류한다:

| 유형 | 불확실성 | 영향도 | 전략적 의미 |
|------|---------|--------|------------|
| **Critical Uncertainties** | 높음 | 높음 | → 시나리오 축 후보 |
| **Key Trends** | 낮음 | 높음 | → 모든 시나리오에 공통 반영 |
| **Monitor** | 높음 | 낮음 | → 와일드카드 가능성 모니터링 |
| **Background** | 낮음 | 낮음 | → 최소한의 모니터링 |

**시나리오 축**은 Critical Uncertainties 중에서 선택한다. 좋은 축은 서로 **독립적**이어야 한다.

아래에서 전기차(EV) 시장 진출을 위한 12개 동인을 분석한다.

````python
# 전기차(EV) 시장의 12개 동인(Driving Forces)
driving_forces = pd.DataFrame({
    'name': [
        'Government Carbon Policy', 'Battery Tech Breakthrough',
        'Oil Price Volatility', 'Consumer Eco-Awareness',
        'Charging Infra Expansion', 'Rare Earth Supply',
        'Autonomous Driving Tech', 'Urban Transport Policy',
        'EV Price Competitiveness', 'Power Grid Stability',
        'Climate Change Acceleration', 'Trade/Tariff Policy'
    ],
    'name_kr': [
        '정부 탄소정책', '배터리 기술 돌파',
        '유가 변동성', '소비자 환경의식',
        '충전 인프라 확충', '희토류 공급 안정성',
        '자율주행 기술', '도시 교통정책 변화',
        'EV 가격 경쟁력', '전력망 안정성',
        '기후변화 가속', '통상/관세 정책'
    ],
    'category': [
        'Political', 'Technological', 'Economic', 'Social',
        'Technological', 'Economic', 'Technological', 'Political',
        'Economic', 'Technological', 'Environmental', 'Political'
    ],
    'uncertainty': [8.5, 9.0, 7.5, 5.0, 6.0, 8.0, 8.5, 7.0, 6.5, 5.5, 4.0, 7.5],
    'impact':      [9.0, 9.5, 7.0, 7.5, 8.5, 8.0, 6.5, 7.5, 9.0, 6.0, 8.0, 6.0]
})

# 4분면 분류
threshold = 6.5
driving_forces['quadrant'] = driving_forces.apply(
    lambda r: 'Critical Uncertainty' if r['uncertainty'] >= threshold and r['impact'] >= threshold
    else ('Key Trend' if r['uncertainty'] < threshold and r['impact'] >= threshold
    else ('Monitor' if r['uncertainty'] >= threshold and r['impact'] < threshold
    else 'Background')), axis=1
)

# STEEP 카테고리별 색상
category_colors = {
    'Political': '#E45756', 'Technological': '#4C78A8',
    'Economic': '#F58518', 'Social': '#72B7B2',
    'Environmental': '#54A24B'
}

fig = go.Figure()

for cat, color in category_colors.items():
    mask = driving_forces['category'] == cat
    df_cat = driving_forces[mask]
    fig.add_trace(go.Scatter(
        x=df_cat['uncertainty'], y=df_cat['impact'],
        mode='markers+text',
        marker=dict(size=14, color=color, line=dict(width=1, color='white')),
        text=df_cat['name'],
        textposition='top center',
        textfont=dict(size=9),
        name=cat,
        hovertemplate='<b>%{text}</b><br>Uncertainty: %{x}<br>Impact: %{y}<extra></extra>'
    ))

# Quadrant lines and labels
fig.add_hline(y=threshold, line_dash="dash", line_color="gray", opacity=0.5)
fig.add_vline(x=threshold, line_dash="dash", line_color="gray", opacity=0.5)

fig.add_annotation(x=8.5, y=9.5, text="Critical Uncertainties<br>(Scenario Axes)", 
                   showarrow=False, font=dict(size=11, color="red"), opacity=0.6)
fig.add_annotation(x=4.5, y=9.0, text="Key Trends", 
                   showarrow=False, font=dict(size=11, color="blue"), opacity=0.6)
fig.add_annotation(x=8.5, y=5.5, text="Monitor", 
                   showarrow=False, font=dict(size=11, color="gray"), opacity=0.6)
fig.add_annotation(x=4.5, y=5.5, text="Background", 
                   showarrow=False, font=dict(size=11, color="gray"), opacity=0.6)

fig.update_layout(
    title='Uncertainty-Impact Matrix: EV Market Driving Forces',
    xaxis_title='Uncertainty (1-10)',
    yaxis_title='Impact (1-10)',
    xaxis=dict(range=[3, 10], dtick=1),
    yaxis=dict(range=[5, 10], dtick=1),
    height=550, width=800,
    legend_title='STEEP Category'
)
fig.show()

# 핵심 불확실성 출력
critical = driving_forces[driving_forces['quadrant'] == 'Critical Uncertainty']
print("\n🔴 핵심 불확실성 (Critical Uncertainties):")
for _, row in critical.iterrows():
    print(f"  • {row['name_kr']} ({row['name']}): 불확실성={row['uncertainty']}, 영향도={row['impact']}")
print(f"\n💡 이 중에서 서로 독립적인 두 동인을 시나리오 축으로 선택한다.")
````

## 2.2 시나리오 구축 5단계 프로세스

Peter Schwartz(1991)가 제안한 시나리오 플래닝 프로세스:

**1단계: 초점 질문(Focal Question) 설정**
- "A사는 향후 5년간 전기차 시장에 어떻게 진출해야 하는가?"
- 구체적이고, 의사결정과 직결되는 질문이어야 한다

**2단계: 동인(Driving Forces) 식별**
- STEEP 프레임워크로 체계적 스캔
- 브레인스토밍, 전문가 인터뷰, 트렌드 분석 병행

**3단계: 불확실성 순위화**
- 불확실성-영향도 매트릭스 작성
- Critical Uncertainties 중 **독립적인** 두 축 선정

**4단계: 시나리오 매트릭스 구성**
- 2×2 매트릭스로 4개 시나리오 구성
- 각 시나리오에 생생한 **서사(narrative)** 부여
- 이름은 직관적이고 기억하기 쉽게

**5단계: 시사점 도출**
- 각 시나리오별 기회와 위협 분석
- **윈드 터널링**: 전략을 각 시나리오에서 테스트
- **조기 경보 지표**: 어떤 시나리오가 전개되는지 모니터링

# Part 3: 2×2 시나리오 매트릭스와 윈드 터널링

## 3.1 시나리오 매트릭스 구성

핵심 불확실성 두 축을 선정했다:
- **축 1 (X축)**: 정부 규제 강도 (약한 규제 ↔ 강한 규제)
- **축 2 (Y축)**: 배터리 기술 발전 속도 (느린 발전 ↔ 빠른 발전)

이로부터 4개의 시나리오가 도출된다:

| 시나리오 | 규제 | 기술 | 핵심 특성 |
|----------|------|------|----------|
| 🟢 Green Acceleration | 강함 | 빠름 | 가장 빠른 전환, 규제와 기술이 상호 가속 |
| 🔵 Regulation-Driven | 강함 | 느림 | 규제 주도 전환, 기술 제약으로 비용 상승 |
| 🟡 Market-Driven | 약함 | 빠름 | 시장 주도 전환, 기술 혁신이 규제를 앞섬 |
| 🔴 Gradual Transition | 약함 | 느림 | 점진적 전환, 내연기관과 장기 공존 |

````python
# 2×2 시나리오 매트릭스 시각화
scenarios = pd.DataFrame({
    'name': ['Green Acceleration', 'Regulation-Driven', 'Market-Driven', 'Gradual Transition'],
    'name_kr': ['녹색 가속', '규제 주도 전환', '시장 주도 전환', '점진적 전환'],
    'regulation': [1, 1, -1, -1],   # 1=Strong, -1=Weak
    'technology': [1, -1, 1, -1],    # 1=Fast, -1=Slow
    'probability': [0.25, 0.20, 0.30, 0.25],
    'color': ['#54A24B', '#4C78A8', '#F58518', '#E45756'],
    'description': [
        'Strong regulation + Fast tech\nRapid EV transition\nGrid parity by 2027',
        'Strong regulation + Slow tech\nPolicy-driven adoption\nHigh subsidy dependence',
        'Weak regulation + Fast tech\nMarket-driven disruption\nTech leaders dominate',
        'Weak regulation + Slow tech\nGradual coexistence\nICE remains competitive'
    ]
})

fig = go.Figure()

# Quadrant background colors (using shapes)
fig.add_shape(type="rect", x0=-1.8, y0=0, x1=0, y1=1.8,
              fillcolor="rgba(245,133,24,0.08)", line_width=0)  # Market-Driven
fig.add_shape(type="rect", x0=0, y0=0, x1=1.8, y1=1.8,
              fillcolor="rgba(84,162,75,0.08)", line_width=0)   # Green Acceleration
fig.add_shape(type="rect", x0=-1.8, y0=-1.8, x1=0, y1=0,
              fillcolor="rgba(228,87,86,0.08)", line_width=0)   # Gradual Transition
fig.add_shape(type="rect", x0=0, y0=-1.8, x1=1.8, y1=0,
              fillcolor="rgba(76,120,168,0.08)", line_width=0)  # Regulation-Driven

for _, s in scenarios.iterrows():
    fig.add_trace(go.Scatter(
        x=[s['regulation'] * 0.8],
        y=[s['technology'] * 0.8],
        mode='markers+text',
        marker=dict(size=50, color=s['color'], opacity=0.7,
                    line=dict(width=2, color='white')),
        text=[f"<b>{s['name']}</b><br>P={s['probability']:.0%}"],
        textposition='middle center',
        textfont=dict(size=10, color='white'),
        hovertext=s['description'],
        hoverinfo='text',
        showlegend=False
    ))

# Axis lines
fig.add_hline(y=0, line_color="gray", line_dash="dash", opacity=0.5)
fig.add_vline(x=0, line_color="gray", line_dash="dash", opacity=0.5)

# Axis labels
fig.add_annotation(x=1.5, y=-1.7, text="<b>Strong Regulation →</b>",
                   showarrow=False, font=dict(size=12))
fig.add_annotation(x=-1.5, y=-1.7, text="<b>← Weak Regulation</b>",
                   showarrow=False, font=dict(size=12))
fig.add_annotation(x=-1.75, y=1.5, text="<b>Fast<br>Tech ↑</b>",
                   showarrow=False, font=dict(size=12))
fig.add_annotation(x=-1.75, y=-1.5, text="<b>↓ Slow<br>Tech</b>",
                   showarrow=False, font=dict(size=12))

fig.update_layout(
    title='2×2 Scenario Matrix: EV Market Entry for Company A',
    xaxis=dict(range=[-1.8, 1.8], showticklabels=False, zeroline=False, showgrid=False),
    yaxis=dict(range=[-1.8, 1.8], showticklabels=False, zeroline=False, showgrid=False),
    height=550, width=700,
    plot_bgcolor='white'
)
fig.show()

print("💡 4개 시나리오가 구성되었다. 각 시나리오의 발생 확률은 전문가 합의에 의해 부여된다.")
print("   가장 높은 확률의 시나리오: Market-Driven (30%)")
````

## 3.2 윈드 터널링 (Wind Tunneling)

윈드 터널링은 **전략을 각 시나리오에서 테스트**하는 기법이다. 항공기를 풍동(wind tunnel)에서 테스트하듯, 전략을 다양한 미래 환경에서 검증한다.

**평가 기준**:
- 각 전략의 **시나리오별 점수** (1-10점): 해당 시나리오에서의 성과
- **기대값(EV)**: EV = Σ P(Sᵢ) × Scoreᵢ — 확률 가중 평균
- **최소값(Min)**: 최악 시나리오에서의 성과 → **강건성(Robustness)** 지표

**전략 선택의 프레임워크**:
- **EV 최대화**: 평균적으로 가장 좋은 전략 (위험 중립적)
- **Min 최대화 (Maximin)**: 최악의 경우를 가장 잘 방어하는 전략 (위험 회피적)
- **균형 접근**: EV와 Min을 동시에 고려

````python
# 윈드 터널링: 4개 전략 × 4개 시나리오
strategies = {
    'Aggressive Entry': [10, 6, 8, 3],
    'Tech Focus':       [9, 5, 10, 4],
    'Gradual Entry':    [6, 7, 7, 8],
    'Hybrid Approach':  [5, 8, 6, 9]
}
scenario_names = ['Green\nAcceleration', 'Regulation\nDriven', 'Market\nDriven', 'Gradual\nTransition']
scenario_probs = [0.25, 0.20, 0.30, 0.25]

# 점수 매트릭스 (Heatmap)
strategy_names = list(strategies.keys())
scores_matrix = np.array(list(strategies.values()))

fig = go.Figure(data=go.Heatmap(
    z=scores_matrix,
    x=scenario_names,
    y=strategy_names,
    text=scores_matrix,
    texttemplate='%{text}',
    textfont=dict(size=16, color='white'),
    colorscale='RdYlGn',
    zmin=1, zmax=10,
    colorbar=dict(title='Score')
))

fig.update_layout(
    title='Wind Tunneling: Strategy × Scenario Score Matrix',
    xaxis_title='Scenario',
    yaxis_title='Strategy',
    height=400, width=700
)
fig.show()

# EV 및 Min 계산
print("\n📊 전략 평가 결과:")
print(f"{'Strategy':<20} {'EV':>6} {'Min':>6} {'Max':>6} {'Range':>6}")
print("-" * 50)

ev_scores = []
min_scores = []
for name, scores in strategies.items():
    ev = sum(p * s for p, s in zip(scenario_probs, scores))
    mn = min(scores)
    mx = max(scores)
    ev_scores.append(ev)
    min_scores.append(mn)
    print(f"{name:<20} {ev:>6.2f} {mn:>6d} {mx:>6d} {mx-mn:>6d}")

print(f"\n🏆 EV 최대 전략: {strategy_names[np.argmax(ev_scores)]} (EV={max(ev_scores):.2f})")
print(f"🛡️ Maximin 전략: {strategy_names[np.argmax(min_scores)]} (Min={max(min_scores)})")
print(f"\n💡 해석: EV 관점에서는 Tech Focus가 우수하지만, 최악의 경우를 고려하면")
print(f"   Gradual Entry가 더 강건한(robust) 전략이다. 의사결정자의 위험 태도에 따라 선택이 달라진다.")
````

````python
# EV vs Min 비교 차트
eval_df = pd.DataFrame({
    'Strategy': list(strategies.keys()),
    'Expected Value': [sum(p*s for p,s in zip(scenario_probs, scores)) for scores in strategies.values()],
    'Minimum Score': [min(scores) for scores in strategies.values()]
})

fig = go.Figure()

fig.add_trace(go.Bar(
    x=eval_df['Strategy'], y=eval_df['Expected Value'],
    name='Expected Value (EV)',
    marker_color='#4C78A8',
    text=eval_df['Expected Value'].round(2),
    textposition='outside'
))

fig.add_trace(go.Bar(
    x=eval_df['Strategy'], y=eval_df['Minimum Score'],
    name='Minimum Score (Robustness)',
    marker_color='#E45756',
    text=eval_df['Minimum Score'],
    textposition='outside'
))

fig.update_layout(
    title='Strategy Evaluation: Expected Value vs Robustness',
    xaxis_title='Strategy',
    yaxis_title='Score',
    barmode='group',
    height=450, width=700,
    yaxis=dict(range=[0, 12]),
    legend=dict(x=0.6, y=0.95)
)
fig.show()

print("💡 EV와 Min 사이의 트레이드오프가 명확하다:")
print("   - Aggressive Entry: 높은 EV(6.80) but 낮은 Min(3) → 고위험-고수익")
print("   - Gradual Entry: 중간 EV(7.00) but 높은 Min(6) → 강건한 전략")
````
